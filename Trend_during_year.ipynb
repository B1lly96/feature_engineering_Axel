```python
def trend_during_year_clay( partita, anno_corrente, anni_precedenti, superficie ) :
    
    import matplotlib.pyplot as plt
    import pandas as pd
    import numpy as np
    
    n1 = partita.split("_")[0]
    n2 = partita.split("_")[1]
    
    
    file_IW = 'DATA/IndianWells_PLACEMENTS_DATA.xlsx'
    file_MIA = 'DATA/Miami_PLACEMENTS_DATA.xlsx'
    file_MON = 'DATA/MonteCarlo_PLACEMENTS_DATA.xlsx'
    file_MAD = 'DATA/Madrid_placements_data.xlsx'
    file_ROM = 'DATA/Rome_PLACEMENTS_DATA.xlsx'
    file_CAN = 'DATA/Canada_PLACEMENTS_DATA.xlsx'
    file_CIN = 'DATA/Cincinnati_PLACEMENTS_DATA.xlsx'
    file_SHA = 'DATA/Shanghai_PLACEMENTS_DATA.xlsx'
    file_PAR = 'DATA/Parigi_1000_PLACEMENTS_DATA.xlsx'

    file_read_IW = pd.read_excel( file_IW )
    dataframe_torneo_IW = pd.DataFrame( file_read_IW )
    
    file_read_MIA = pd.read_excel( file_MIA )
    dataframe_torneo_MIA = pd.DataFrame( file_read_MIA )
    
    file_read_MON = pd.read_excel( file_MON )
    dataframe_torneo_MON = pd.DataFrame( file_read_MON )
    
    file_read_MAD = pd.read_excel( file_MAD )
    dataframe_torneo_MAD = pd.DataFrame( file_read_MAD )
    
    file_read_ROM = pd.read_excel( file_ROM )
    dataframe_torneo_ROM = pd.DataFrame( file_read_ROM )
    
    file_read_CAN = pd.read_excel( file_CAN )
    dataframe_torneo_CAN = pd.DataFrame( file_read_CAN )
    
    file_read_CIN = pd.read_excel( file_CIN )
    dataframe_torneo_CIN = pd.DataFrame( file_read_CIN )
    
    file_read_SHA = pd.read_excel( file_SHA )
    dataframe_torneo_SHA = pd.DataFrame( file_read_SHA )
    
    file_read_PAR = pd.read_excel( file_PAR )
    dataframe_torneo_PAR = pd.DataFrame( file_read_PAR )
    
    n1 = partita.split("_")[0]
    n2 = partita.split("_")[1]
    dict_placements = {1:7,
                       2:6,
                       3:5, 4:5,
                       5:4, 6:4, 7:4, 8:4,
                       9:3, 10:3, 11:3, 12:3, 13:3, 14:3, 15:3, 16:3,
                       17:2, 18:2, 19:2, 20:2, 21:2, 22:2, 23:2, 24:2, 
                       25:2, 26:2, 27:2, 28:2, 29:2, 30:2, 31:2, 32:2 }
    dict_anni = {'2014':0,
                 '2015':1,
                 '2016':2,
                 '2017':3,
                 '2018':4,
                 '2019':5,
                 '2020':6,
                 '2021':7 }
    
    years = dataframe_torneo_IW.iloc[:,0]
    idx_year = 0
    for idx, v in enumerate ( years ) :
        if int( v )==anno_corrente :
            idx_year = idx_year+idx-anni_precedenti 
    
    # N1
    n1_place_MON = 0
    for i, e in enumerate ( list( dataframe_torneo_MON.iloc[idx_year, :] ) ) :
        if n1==e :
            n1_place_MON = n1_place_MON+i
    
    n1_place_MAD = 0
    for i, e in enumerate ( list( dataframe_torneo_MAD.iloc[idx_year, :] ) ) :
        if n1==e :
            n1_place_MAD = n1_place_MAD+i
    
    n1_place_ROM = 0
    for i, e in enumerate ( list( dataframe_torneo_ROM.iloc[idx_year, :] ) ) :
        if n1==e :
            n1_place_ROM = n1_place_ROM+i
    
    # N2
    n2_place_MON = 0
    for i, e in enumerate ( list( dataframe_torneo_MON.iloc[idx_year, :] ) ) :
        if n2==e :
            n2_place_MON = n2_place_MON+i
    
    n2_place_MAD = 0
    for i, e in enumerate ( list( dataframe_torneo_MAD.iloc[idx_year, :] ) ) :
        if n2==e :
            n2_place_MAD = n2_place_MAD+i
    
    n2_place_ROM = 0
    for i, e in enumerate ( list( dataframe_torneo_ROM.iloc[idx_year, :] ) ) :
        if n2==e :
            n2_place_ROM = n2_place_ROM+i
    
    n1_clay = [ n1_place_MON, n1_place_MAD, n1_place_ROM ]
    n2_clay = [ n2_place_MON, n2_place_MAD, n2_place_ROM ]
    
    n1_clay = pd.Series( n1_clay )
    n2_clay = pd.Series( n2_clay )
    
    n1_place = n1_clay.map( dict_placements )
    n2_place = n2_clay.map( dict_placements )
    
    n1_place = n1_place.fillna( -1 )
    
    n2_place = n2_place.fillna( -1 )
    
    
    passi = [1,2,3]
    
    title = "MASTER 1000 "+superficie+" "+str( anno_corrente-anni_precedenti )
    plt.figure( figsize=( 7,7 ) )
    plt.plot( passi, n1_place, '-o', label=n1 )
    plt.plot( passi, n2_place, '-o', label=n2 )
    plt.xticks([1,2,3],["Montecarlo","Madrid","Roma"] )
    plt.yticks([-1,2,3,4,5,6,7],['NaN','32s','OTT','QRT','SF','F','W'] )
    plt.title( title )
    plt.legend()
    plt.grid()
    plt.show()
    
    
    
            
    
    
    
    
    
            
```

```python
def masters_rosso_last_3_2_1_0( partita, anno_corrente, superficie) :
    trend_during_year_clay( partita, anno_corrente, 3, superficie )
    trend_during_year_clay( partita, anno_corrente, 2, superficie )
    trend_during_year_clay( partita, anno_corrente, 1, superficie )
    trend_during_year_clay( partita, anno_corrente, 0, superficie )
```
