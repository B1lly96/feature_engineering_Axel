```python
def diff_somm( lista1, lista2, nome_diff, nome_plus ) :
    for idx, v in enumerate ( lista1 ) :
        diff = v - lista2[idx]
        plus = v + lista2[idx]
        nome_diff.append( diff )
        nome_plus.append( plus )
```
